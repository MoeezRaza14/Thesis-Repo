# **Fusing the Data**

```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

# ---- CONFIGURATION ----
INPUT_CSV = '/content/drive/MyDrive/Synthetic Data/Synthetic Stress DRYAD data.csv'          # Path to your full dataset
OUTPUT_CSV = 'balanced_subset_dryad_stress.csv'   # Where to save the subset
TARGET_COL = 'Stress'      # Your label column name
TOTAL_SAMPLES = 6897
RANDOM_STATE = 42

# ---- LOAD DATA ----
df = pd.read_csv(INPUT_CSV)

# ---- STRATIFIED SPLIT ----
sss = StratifiedShuffleSplit(
    n_splits=1,
    train_size=TOTAL_SAMPLES,
    random_state=RANDOM_STATE
)
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]

train_idx, _ = next(sss.split(X, y))
subset = df.iloc[train_idx].reset_index(drop=True)

# ---- VERIFY DISTRIBUTION ----
dist = subset[TARGET_COL].value_counts().sort_index()
print("Label counts in subset:\n", dist)

# ---- SAVE SUBSET ----
subset.to_csv(OUTPUT_CSV, index=False)
print(f"\nSaved {len(subset)}‑row subset to {OUTPUT_CSV}")
```
